## 1. Temporal Analysis of Mental Health Disorders


a) Are there trends of increase or decrease in the rates of depression or anxiety disorders in specific years?

```python
%%bigquery
WITH mental_health_data AS (
    SELECT
        Year,
        FORMAT('%.4f', AVG(CAST(Depression_Percent AS FLOAT64))) AS avg_depression_rate,
        FORMAT('%.4f', AVG(CAST(Anxiety_disorders_Percent AS FLOAT64))) AS avg_anxiety_rate,
        FORMAT('%.4f', AVG(CAST(Depression_Percent AS FLOAT64)) - LAG(AVG(CAST(Depression_Percent AS FLOAT64))) OVER (ORDER BY Year)) AS depression_change,
        FORMAT('%.4f', AVG(CAST(Anxiety_disorders_Percent AS FLOAT64)) - LAG(AVG(CAST(Anxiety_disorders_Percent AS FLOAT64))) OVER (ORDER BY Year)) AS anxiety_change
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2006 AND 2016
    GROUP BY
        Year
)

SELECT *
FROM mental_health_data
WHERE EXTRACT(YEAR FROM Year) >= 2007
ORDER BY Year;
```

**Summary**

The data showcases trends in average depression and anxiety rates from 2007 to 2016. The average depression rates show a gradual decline from 3.4925 in 2007 to 3.4544 in 2016, while anxiety rates remain relatively stable, fluctuating around 3.996 to 3.997. Depression rate changes exhibit a consistent decrease year-over-year, with the largest drop of -0.0108 occurring in 2008. Anxiety rate changes, on the other hand, show smaller fluctuations, with minimal positive or negative variations, indicating near stability. Overall, depression rates steadily decrease, while anxiety rates remain relatively unchanged during the period.

c) Is there a difference in the year-by-year changes in prevalence between depression and anxiety disorders?

```python
%%bigquery
WITH mental_health_changes AS (
    SELECT
        Entity AS Country_Name,
        EXTRACT(YEAR FROM Year) AS Year,
        AVG(CAST(Depression_Percent AS FLOAT64)) AS avg_depression_rate,
        AVG(CAST(Anxiety_disorders_Percent AS FLOAT64)) AS avg_anxiety_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2006 AND 2016
    GROUP BY
        Entity, EXTRACT(YEAR FROM Year)
),

yearly_changes AS (
    SELECT
        Country_Name,
        Year,
        avg_depression_rate,
        avg_anxiety_rate,
        LAG(avg_depression_rate) OVER (PARTITION BY Country_Name ORDER BY Year) AS prev_depression_rate,
        LAG(avg_anxiety_rate) OVER (PARTITION BY Country_Name ORDER BY Year) AS prev_anxiety_rate
    FROM
        mental_health_changes
),

percentage_changes AS (
    SELECT
        Country_Name,
        Year,
        (avg_depression_rate - prev_depression_rate) / prev_depression_rate * 100 AS depression_change_percentage,
        (avg_anxiety_rate - prev_anxiety_rate) / prev_anxiety_rate * 100 AS anxiety_change_percentage,
        ABS((avg_depression_rate - prev_depression_rate) / prev_depression_rate * 100 -
            (avg_anxiety_rate - prev_anxiety_rate) / prev_anxiety_rate * 100) AS change_difference
    FROM
        yearly_changes
    WHERE
        prev_depression_rate IS NOT NULL
        AND prev_anxiety_rate IS NOT NULL
)

SELECT
    Year,
    FORMAT('%.4f', AVG(depression_change_percentage)) AS avg_depression_change,
    FORMAT('%.4f', AVG(anxiety_change_percentage)) AS avg_anxiety_change,
    FORMAT('%.4f', AVG(change_difference)) AS avg_change_difference
FROM
    percentage_changes
WHERE
    Year BETWEEN 2007 AND 2016
GROUP BY
    Year
ORDER BY
    Year;
```

**Summary**

The year-by-year changes in the prevalence of depression and anxiety disorders reveal distinct trends between 2007 and 2016. The average percentage change in depression rates consistently decreased, starting at -0.2797 in 2007 and declining to -0.0037 in 2016, indicating a significant reduction in depression prevalence over time. Conversely, the average percentage change in anxiety rates remained relatively stable, with minor fluctuations around positive values, such as 0.0234 in 2007 and 0.0064 in 2016, suggesting minimal increases in anxiety prevalence. The difference in change rates between depression and anxiety narrowed over the years, with the average change difference decreasing from 0.4170 in 2007 to 0.2626 in 2016, reflecting a convergence in their year-over-year prevalence trends. Overall, depression rates showed a consistent downward trend, while anxiety rates exhibited stability with slight upward changes.

## 2. Demographical Analysis of Mental Health Disorders


a) Are there any trends where depression or anxiety rates increase more rapidly in certain countries compared to others?

```python
%%bigquery
WITH mental_health_changes AS (
    SELECT
        Entity AS Country_Name,
        EXTRACT(YEAR FROM Year) AS Year,
        AVG(CAST(Depression_Percent AS FLOAT64)) AS avg_depression_rate,
        AVG(CAST(Anxiety_disorders_Percent AS FLOAT64)) AS avg_anxiety_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016
        AND Entity NOT IN ('World', 'North America', 'Oceania', 'Sub-Saharan Africa',
                           'High SDI', 'Low SDI', 'Middle SDI', 'Latin America and Caribbean',
                           'Southeast Asia', 'Eastern Europe', 'Western Europe')
    GROUP BY
        Entity, EXTRACT(YEAR FROM Year)
),

yearly_changes AS (
    SELECT
        Country_Name,
        Year,
        avg_depression_rate,
        avg_anxiety_rate,
        LAG(avg_depression_rate) OVER (PARTITION BY Country_Name ORDER BY Year) AS prev_depression_rate,
        LAG(avg_anxiety_rate) OVER (PARTITION BY Country_Name ORDER BY Year) AS prev_anxiety_rate
    FROM
        mental_health_changes
),

percentage_changes AS (
    SELECT
        Country_Name,
        Year,
        (avg_depression_rate - prev_depression_rate) / prev_depression_rate * 100 AS depression_change_percentage,
        (avg_anxiety_rate - prev_anxiety_rate) / prev_anxiety_rate * 100 AS anxiety_change_percentage
    FROM
        yearly_changes
    WHERE
        prev_depression_rate IS NOT NULL
        AND prev_anxiety_rate IS NOT NULL
        AND Year IN (2007, 2016)
)

SELECT
    Country_Name,
    FORMAT("%.2f", AVG(depression_change_percentage)) AS avg_depression_change_rate,
    FORMAT("%.2f", AVG(anxiety_change_percentage)) AS avg_anxiety_change_rate,
    FORMAT("%.2f", (AVG(depression_change_percentage) + AVG(anxiety_change_percentage)) / 2) AS total_change_rate
FROM
    percentage_changes
GROUP BY
    Country_Name
ORDER BY
    CAST(total_change_rate AS FLOAT64) DESC
LIMIT 10;
```

**Summary:**

The query identifies countries where depression and anxiety rates have increased most significantly between 2007 and 2016. Japan leads the list with an average depression change rate of 1.18% and a modest anxiety change rate of 0.07%, resulting in a total change rate of 0.62%. Other notable countries include Iraq (total change rate 0.55%) and Zambia (total change rate 0.55%), where depression and anxiety rates have also seen notable shifts. These results highlight countries experiencing relatively rapid increases in mental health disorder rates compared to others.

b) Which countries have highest difference in Anxiety rates between Men and Women?

```python
%%bigquery
WITH avg_anxiety AS (
    SELECT
        Entity AS country,
        ROUND(AVG(CAST(`Prevalence - Anxiety disorders - Sex: Female - Age: Age-standardized _Percent_` AS FLOAT64)), 2) AS female_anxiety,
        ROUND(AVG(CAST(`Prevalence - Anxiety disorders - Sex: Male - Age: Age-standardized _Percent_` AS FLOAT64)), 2) AS male_anxiety
    FROM
        `ba775-team-3.final_cleaned_dataset.All_Prevalance`
    WHERE
        Year BETWEEN 2007 AND 2016
        AND `Prevalence - Anxiety disorders - Sex: Female - Age: Age-standardized _Percent_` IS NOT NULL
        AND `Prevalence - Anxiety disorders - Sex: Male - Age: Age-standardized _Percent_` IS NOT NULL
        AND Entity NOT IN ('World', 'North America', 'Oceania', 'Sub-Saharan Africa',
                           'High SDI', 'Low SDI', 'Middle SDI', 'Latin America and Caribbean',
                           'Southeast Asia', 'Eastern Europe', 'Western Europe')
    GROUP BY
        Entity
)
SELECT
    country,
    female_anxiety,
    male_anxiety,
    ROUND(ABS(male_anxiety - female_anxiety), 2) AS disparity
FROM
    avg_anxiety
ORDER BY
    disparity DESC
LIMIT 10;
```

**Summary**

The table highlights the countries with the largest disparity in anxiety prevalence rates between women and men from 2007 to 2016. Brazil shows the highest disparity with female anxiety rates at 11.14% and male anxiety rates at 5.54%, resulting in a difference of 5.60%. Other notable countries include Chile (disparity of 5.40%), Spain (5.34%), and France (4.78%). These findings emphasize significant gender disparities in anxiety prevalence, with women consistently exhibiting higher rates than men across all listed countries.

c) Which countries have highest difference in Depression rates between Men and Women?

```python
%%bigquery
WITH average_depression AS (
    SELECT
        Entity AS country,
        ROUND(AVG(CAST(`Prevalence - Depressive disorders - Sex: Female - Age: Age-standardized _Percent_` AS FLOAT64)), 2) AS female_depression,
        ROUND(AVG(CAST(`Prevalence - Depressive disorders - Sex: Male - Age: Age-standardized _Percent_` AS FLOAT64)), 2) AS male_depression
    FROM `ba775-team-3.final_cleaned_dataset.All_Prevalance`
    WHERE Year BETWEEN 2007 AND 2016
        AND `Prevalence - Depressive disorders - Sex: Female - Age: Age-standardized _Percent_` IS NOT NULL
        AND `Prevalence - Depressive disorders - Sex: Male - Age: Age-standardized _Percent_` IS NOT NULL
        AND Entity NOT IN ('World', 'North America', 'Oceania', 'Sub-Saharan Africa',
                           'High SDI', 'Low SDI', 'Middle SDI', 'Latin America and Caribbean',
                           'Southeast Asia', 'Eastern Europe', 'Western Europe')
    GROUP BY country
)
SELECT
    country,
    female_depression,
    male_depression,
    ROUND(ABS(male_depression - female_depression), 2) AS disparity
FROM average_depression
ORDER BY disparity DESC
LIMIT 10;
```

**Summary**

The analysis identifies the top 10 countries with the highest disparities in depression rates between women and men from 2007 to 2016. Gambia leads with a significant gap of 4.76%, followed by Liberia (4.42%) and Greenland (3.16%).
```python
%%bigquery
WITH happiness_data AS (
    SELECT
        `Country name` AS Country_Name,
        Year,
        `Life Ladder` AS Happiness_Score
    FROM
        `ba775-team-3.final_cleaned_dataset.World Happiness Report 2005_2021`
    WHERE
        Year BETWEEN 2007 AND 2016  -- Only consider years 2007 to 2016
),

depression_data AS (
    SELECT
        Entity AS Country_Name,
        EXTRACT(YEAR FROM Year) AS Year,
        AVG(CAST(Depression_Percent AS FLOAT64)) AS avg_depression_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016  -- Only consider years 2007 to 2016
    GROUP BY
        Country_Name, Year
),

combined_data AS (
    SELECT
        h.Country_Name,
        h.Year,
        h.Happiness_Score,
        d.avg_depression_rate
    FROM
        happiness_data h
    JOIN
        depression_data d
    ON
        h.Country_Name = d.Country_Name
        AND h.Year = d.Year
),

average_data AS (
    SELECT
        Country_Name,
        FORMAT('%.2f', AVG(Happiness_Score)) AS avg_happiness_score,
        FORMAT('%.2f', AVG(avg_depression_rate)) AS avg_depression_rate
    FROM
        combined_data
    GROUP BY
        Country_Name
),

ranked_data AS (
    SELECT
        Country_Name,
        avg_happiness_score,
        avg_depression_rate,
        ROW_NUMBER() OVER (ORDER BY CAST(avg_happiness_score AS FLOAT64) DESC) AS happiness_rank,
        ROW_NUMBER() OVER (ORDER BY CAST(avg_depression_rate AS FLOAT64) ASC) AS depression_rank
    FROM
        average_data
),

top_happiness AS (
    SELECT *
    FROM ranked_data
    ORDER BY happiness_rank ASC
    LIMIT 10
),

bottom_happiness AS (
    SELECT *
    FROM ranked_data
    ORDER BY happiness_rank DESC
    LIMIT 10
)

SELECT *
FROM top_happiness
UNION ALL
SELECT *
FROM bottom_happiness
ORDER BY happiness_rank;
```

**Summary**

The results suggest a relationship between higher happiness scores and lower depression rates. Countries like Denmark, Norway, and Switzerland, which rank highest in happiness (7.67, 7.59, and 7.56, respectively), generally exhibit lower average depression rates (3.28, 3.65, and 3.75, respectively). Conversely, countries with lower happiness scores, such as Togo (3.25), Burundi (3.49), and the Central African Republic (3.52), have higher depression rates (3.66, 3.76, and 4.22, respectively). While there are exceptions, the data indicates a trend where countries with greater happiness tend to experience lower rates of depression, supporting the hypothesis that these metrics are inversely related.

### 3-3. Suicide Rates

a) When viewed on a national level, what impact does depression have on the suicide rate?

```python
%%bigquery
WITH FilteredData AS (
    SELECT
        Entity,
        ROUND(AVG(Depression_Percent), 2) AS avg_depression_percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016
    GROUP BY
        Entity
),
SuicideData AS (
    SELECT
        Entity,
        ROUND(AVG(Age_standardized_death_rate_from_self_harm_both_sexes), 2) AS avg_suicide_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Suicide_Rate`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016
        AND Age_standardized_death_rate_from_self_harm_both_sexes IS NOT NULL
    GROUP BY
        Entity
)
SELECT
    fd.Entity,
    fd.avg_depression_percent,
    sd.avg_suicide_rate
FROM
    FilteredData fd
LEFT JOIN
    SuicideData sd
ON
    fd.Entity = sd.Entity
WHERE
    sd.avg_suicide_rate IS NOT NULL
ORDER BY
    fd.avg_depression_percent DESC
LIMIT 10;
```

b) What is the correlation between Depression and Suicide Rates?

```python
%%bigquery
WITH FilteredData AS (
    SELECT
        Entity,
        ROUND(AVG(Depression_Percent), 2) AS avg_depression_percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016
    GROUP BY
        Entity
),
SuicideData AS (
    SELECT
        Entity,
        ROUND(AVG(Age_standardized_death_rate_from_self_harm_both_sexes), 2) AS avg_suicide_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Suicide_Rate`
    WHERE
        EXTRACT(YEAR FROM Year) BETWEEN 2007 AND 2016
        AND Age_standardized_death_rate_from_self_harm_both_sexes IS NOT NULL
    GROUP BY
        Entity
),
CombinedData AS (
    SELECT
        fd.Entity,
        LOG(fd.avg_depression_percent + 1) AS log_avg_depression_percent,
        LOG(sd.avg_suicide_rate + 1) AS log_avg_suicide_rate
    FROM
        FilteredData fd
    LEFT JOIN
        SuicideData sd
    ON
        fd.Entity = sd.Entity
    WHERE
        sd.avg_suicide_rate IS NOT NULL
)

-- Calculate correlation with log-transformed data
SELECT
    ROUND(CORR(log_avg_depression_percent, log_avg_suicide_rate), 4) AS log_correlation_coefficient
FROM
    CombinedData;
```

**Summary**:

The analysis demonstrates a moderate positive relationship between national depression rates and suicide rates, with a log-transformed correlation coefficient of 0.3284. Countries with higher average depression rates, such as Morocco (5.64%) and Lesotho (5.60%), exhibit varying suicide rates, with Lesotho having a notably high suicide rate of 30.76 per 100,000 people, while Morocco's rate is significantly lower at 3.95. This suggests that while depression is a contributing factor to suicide rates, other social, cultural, and economic factors likely influence the variability in the impact of depression on suicide across countries.
```python
%%bigquery
WITH DisorderData AS (
    SELECT
        Entity,
        "Schizophrenia" AS Disorder,
        ROUND(AVG(Schizophrenia_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Bipolar Disorder" AS Disorder,
        ROUND(AVG(Bipolar_disorder_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Eating Disorders" AS Disorder,
        ROUND(AVG(Eating_disorders_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Anxiety Disorders" AS Disorder,
        ROUND(AVG(Anxiety_disorders_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Drug Use Disorders" AS Disorder,
        ROUND(AVG(Drug_use_disorders_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Depression" AS Disorder,
        ROUND(AVG(Depression_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
    UNION ALL
    SELECT
        Entity,
        "Alcohol Use Disorders" AS Disorder,
        ROUND(AVG(Alcohol_use_disorders_Percent), 2) AS Avg_Disorder_Percent
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
    WHERE
        Entity IN ("Russia", "South Korea", "South Africa", "Japan", "India",
                   "France", "United States", "Ethiopia", "Kenya", "Germany")
    GROUP BY
        Entity
)
, RankedDisorders AS (
    SELECT
        Entity,
        Disorder,
        MAX(Avg_Disorder_Percent) AS Max_Disorder_Percent,
        ROW_NUMBER() OVER (
            PARTITION BY Entity
            ORDER BY MAX(Avg_Disorder_Percent) DESC
        ) AS Rank
    FROM
        DisorderData
    GROUP BY
        Entity, Disorder
)
SELECT
    Entity,
    Disorder,
    Max_Disorder_Percent
FROM
    RankedDisorders
WHERE
    Rank <= 3 -- Only top 3 disorders per country
ORDER BY
    CASE
        WHEN Entity = "Russia" THEN 1
        WHEN Entity = "South Korea" THEN 2
        WHEN Entity = "South Africa" THEN 3
        WHEN Entity = "Japan" THEN 4
        WHEN Entity = "India" THEN 5
        WHEN Entity = "France" THEN 6
        WHEN Entity = "United States" THEN 7
        WHEN Entity = "Ethiopia" THEN 8
        WHEN Entity = "Kenya" THEN 9
        WHEN Entity = "Germany" THEN 10
        ELSE 11
    END,
    Rank;
```

### 3-4. Unemployment

a) What is the overall correlation between the unemployment rate and mental health disorders?

```python
%%bigquery
WITH unemployment_data AS (
    SELECT
        Country_Name AS country,
        Year AS year,
        Code AS code,
        Unemployment_percent_of_total_labor_force AS unemployment_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Unemployment_Rate`
),

mental_health_data AS (
    SELECT
        Code AS code,
        EXTRACT(YEAR FROM Year) AS year,
        Depression_percent AS depression_rate,
        Anxiety_disorders_Percent AS anxiety_rate,
        Drug_use_disorders_Percent AS drug_use_rate,
        Bipolar_disorder_percent AS bipolar_rate,
        Schizophrenia_percent AS schizophrenia_rate,
        Eating_disorders_percent AS eating_disorder_rate,
        Alcohol_use_disorders_Percent AS alcohol_use_rate
    FROM
        `ba775-team-3.final_cleaned_dataset.Mental_health_Depression_disorder_Data`
),

correlation_data AS (
    SELECT
        CORR(u.unemployment_rate, m.depression_rate) AS corr_depression,
        CORR(u.unemployment_rate, m.anxiety_rate) AS corr_anxiety,
        CORR(u.unemployment_rate, m.drug_use_rate) AS corr_substance_use,
        CORR(u.unemployment_rate, m.bipolar_rate) AS corr_bipolar,
        CORR(u.unemployment_rate, m.schizophrenia_rate) AS corr_schizophrenia,
        CORR(u.unemployment_rate, m.eating_disorder_rate) AS corr_eating_disorder,
        CORR(u.unemployment_rate, m.alcohol_use_rate) AS corr_alcohol_use
    FROM
        unemployment_data u
    JOIN
        mental_health_data m
    ON
        u.code = m.code
        AND u.year = m.year
)

SELECT
    ROUND(corr_depression, 4) AS corr_depression,
    ROUND(corr_anxiety, 4) AS corr_anxiety,
    ROUND(corr_substance_use, 4) AS corr_substance_use,
    ROUND(corr_bipolar, 4) AS corr_bipolar,
    ROUND(corr_schizophrenia, 4) AS corr_schizophrenia,
    ROUND(corr_eating_disorder, 4) AS corr_eating_disorder,
    ROUND(corr_alcohol_use, 4) AS corr_alcohol_use,
    ROUND((corr_depression + corr_anxiety + corr_substance_use + corr_bipolar + corr_schizophrenia + corr_eating_disorder + corr_alcohol_use) / 7, 4) AS avg_correlation
FROM
    correlation_data;
```

Substance use disorders (0.008) and bipolar disorder (0.0693) show slightly positive correlations, while schizophrenia (-0.2003) and eating disorders (-0.0822) demonstrate moderate negative correlations, suggesting that higher unemployment rates may correspond to lower prevalence of these disorders. Alcohol use disorders have a near-zero correlation (0.0189), showing minimal impact from unemployment rates. The average correlation across all disorders is -0.0307, highlighting an overall weak and inconsistent relationship between unemployment and mental health disorders.

- **Collaboration Issues**:  
   - Tools like Tableau and Vertex AI lacked the ability for simultaneous editing, making it hard for team members to collaborate efficiently.  
   - Sequential workflows delayed progress and complicated merging contributions.
   
- **SQL Limitations**:  
   - While SQL was effective for basic queries, it fell short for advanced analyses like statistical modeling.  
   - Tasks often required exporting data to tools like Python or R, adding complexity and risk of errors.
   
- **Data Cleaning Inefficiencies**:  
   - Preprocessing data using SQL was cumbersome, especially for handling null values and standardizing formats.  
   - SQL’s limitations made certain cleaning tasks time-intensive compared to using specialized tools.

- **Multiple Table Complexity**:  
   - The dataset contained over 10 tables, requiring repeated joins for integration and analysis.  
   - This added time to data preparation and increased the potential for errors, slowing down exploratory work.
